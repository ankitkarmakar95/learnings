```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
```

## **This code is to make a decision tree which can minimise the credit risk for a bank**

```python
creditrisk=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/credit_risk.csv')
```

```python
creditrisk.head()
```

```python
creditrisk.info()
```

```python
X=creditrisk[creditrisk.columns.drop("class")]
Y=creditrisk["class"]
```

```python
print("X shape =", X.shape)
print("Y shape =", Y.shape)
```

```python
creditrisk_encoded = pd.get_dummies(X)
print("Total number of predictors after encoding = ", len(creditrisk_encoded.columns))
creditrisk_encoded.columns
```

```python
X_train,X_test,y_train,y_test = train_test_split(creditrisk_encoded, Y,test_size=0.15,random_state=100) 
print("Shape of X_train and y_train are:", X_train.shape, "and", y_train.shape, " respectively")
print("Shape of X_test and y_test are:", X_test.shape, "and", y_test.shape, " respectively")
```

```python
model = DecisionTreeClassifier(random_state = 1)
model.fit(X_train,y_train)
```

```python
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

```python
dot_data = export_graphviz(model, out_file=None,
                          feature_names=creditrisk_encoded.columns,  
                          class_names=model.classes_,  
                          ) 
```

```python
graph = graphviz.Source(dot_data) 
graph
```

```python
train_accuracy = model.score(X_train,y_train)
print("Accuracy of the model on train data = ",train_accuracy)
```

```python
test_accuracy = model.score(X_test,y_test)
print("Accuracy of the model on test data = ",test_accuracy)
```

```python
model1 = DecisionTreeClassifier(min_samples_split=5,min_impurity_decrease=0.005)
model1.fit(X_train,y_train)
print("train_accuracy = ", model1.score(X_train,y_train))
print("test_accuracy = ", model1.score(X_test,y_test))
```
