Feature Engineering

````python
main=housing[['Price','Date',"No_of_bedrooms","Bathrooms", "Living_sqft", "Floors","Waterfront", "View", "Condition", "Grade","Zip"]]
````

````python
main
````

````python
main["Year"]=main['Date'].str[0:4]
main["Month"]=main['Date'].str[4:6]
````

````python
main=main.drop(columns=['Date'])
````

````python
#encoding catagorical values
````

````python
main
````

````python
cat_ftrs=["Waterfront", "View","Condition", "Grade","Zip",'Month','Year']
main=pd.get_dummies(main,columns=cat_ftrs)
print(main.columns)
````

To make the model more reasonable, we need to normalise the features with continous numerical values so that each of these features span accross a common range of value, so that they don't influence more than the other features of the model.

These column values to be normalised.
price, bedrooms, bathrooms, spft_living and floors.

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(main[['Price','No_of_bedrooms','Bathrooms','Living_sqft','Floors']])
main_norm=scaler.transform(main[['Price','No_of_bedrooms','Bathrooms','Living_sqft','Floors']])
main_norm
````

````python
main_norm_df=pd. DataFrame(main_norm, columns=['Price','No_of_bedrooms','Bathrooms','Living_sqft','Floors'])
main_norm_df=main_norm_df.join(main[main.columns.drop(['Price','No_of_bedrooms','Bathrooms','Living_sqft','Floors'])])
````

Model Building

````python
Y = main_norm_df['Price']
X = main_norm_df[main_norm_df.columns.drop(['Price'])]
X.shape
````

Spliting test & train data

````python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15, random_state=100)
X_train.shape , X_test.shape
````

Building the model

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
````

Testing the model

````python
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
````

# Evaluation using root mean square

````python
from sklearn.metrics import mean_squared_error
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.5
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.5
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
````
